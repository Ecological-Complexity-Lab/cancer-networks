```python
import numpy as np
import pandas as pd
import json

organs = ["breast", "colon", "head", "kidneyc", "kidneyp", "liver", "lunga", "lungs", "prostate", "stomach", "thyroid","uterine"]
with open('node_idx.json', 'r') as fp:
    node_name_dict = json.load(fp)

genes = ['HSPA9', 'HSPD1', 'HSPE1', 'SPG7', 'CLPP', 'DNAJA3', 'LONP1',
       'TRAP1', 'YME1L1', 'CLPX', 'AFG3L2', 'HTRA2', 'DNAJC19', 'GRPEL2',
       'HSCB']

tbd = pd.read_csv("../data/Uterine_Corpus_Endometrial_Carcinoma.csv")
protein = list(tbd.columns[1:].values)
data_path = r"../data/"
result_path = r"../results/"
projected_name = 'community_projected_breast_colon_head_kidneyc_kidneyp_liver_lunga_lungs_prostate_stomach_thyroid_uterine_'
bipartite_name = 'community_breast_colon_head_kidneyc_kidneyp_liver_lunga_lungs_prostate_stomach_thyroid_uterine_'
```

```python

likelihood = []
membership = []

for k in range(2,16):
    
    temp = []
    with open(result_path+projected_name+'_u_K'+str(k)+'.dat') as f:
        first_line = f.readline()
        lines = (line for line in f if not line.startswith('#'))
        uk5 = np.loadtxt(lines, delimiter=' ', skiprows = 0 )
        
    uk5= uk5[:,1:]
    uk5 = np.array(uk5)
    likelihood.append(first_line)
    
    for u in uk5:
        v = u.tolist()
        com = v.index(max(v))+1
        temp.append(com)
    membership.append(temp)
    
    #print(uk5)
    #print(first_line)
```

```python
projected_membership = np.array(membership).transpose()
columns = ["#community_" + str(k) for k in range(2,16)]
index = genes
df1 = pd.DataFrame(projected_membership, index, columns)
df1.to_csv('projected_membership.csv')  
```
